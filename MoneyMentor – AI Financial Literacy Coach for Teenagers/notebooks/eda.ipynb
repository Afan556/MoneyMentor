```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv(r'student_spending.csv')
print(df.head(5))

spending_catagories=[
    'tuition', 'housing', 'food', 'transportation', 'books_supplies',
    'entertainment', 'personal_care', 'technology', 'health_wellness',
    'miscellaneous'
]

total_spending_per_catagory= df[spending_catagories].sum()
total_spending_per_catagory=total_spending_per_catagory.sort_values(ascending=False)

plt.figure(figsize=(12, 6))
plt.bar(total_spending_per_catagory.index, total_spending_per_catagory.values, color='skyblue')
plt.title('Total Spending per Category')
plt.xlabel('Spending Category')
plt.ylabel('Total Spending ($)')
plt.xticks(rotation=45,ha='right')
plt.tight_layout()
plt.show()
plt.savefig('total_spending_per_catagory.png')
print("total_spending_per_catagory.png saved successfully.")
```

```python
income_sources = ['monthly_income', 'financial_aid']

# Calculate total income from each source
total_income = df[income_sources].sum()

# Create the pie chart for income sources
plt.figure(figsize=(8, 8))
plt.pie(total_income, labels=total_income.index, autopct='%1.1f%%', startangle=90, colors=['gold', 'lightgreen'])
plt.title('Proportion of Income Sources')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.savefig('income_sources_pie_chart.png')
print("Pie chart for income sources saved as 'income_sources_pie_chart.png'")
```

```python
import pandas as pd

df_financial_literacy = pd.read_csv('Financial literacy.csv')

# Display the first 5 rows
print("Financial Literacy Data - Head:")
print(df_financial_literacy.head().to_markdown(index=False, numalign="left", stralign="left"))

# Display column information
print("\nFinancial Literacy Data - Info:")
print(df_financial_literacy.info())
```

```python
# Create a histogram for 'TOTAL' score distribution
plt.figure(figsize=(10, 6))
plt.hist(df_financial_literacy['TOTAL'], bins=10, edgecolor='black', color='skyblue')
plt.title('Distribution of Financial Literacy Scores')
plt.xlabel('Financial Literacy Score (TOTAL)')
plt.ylabel('Number of Respondents')
plt.grid(axis='y', alpha=0.75)
plt.savefig('financial_literacy_score_distribution_histogram.png')
print("Histogram for score distribution saved as 'financial_literacy_score_distribution_histogram.png'")
```

```python
# Box plot: Score by gender
plt.figure(figsize=(8, 6))
sns.boxplot(x='P3', y='TOTAL', data=df_financial_literacy, palette='pastel')
plt.title('Financial Literacy Score by Gender')
plt.xlabel('Gender')
plt.ylabel('Financial Literacy Score')
plt.savefig('financial_literacy_score_by_gender_boxplot.png')
print("Box plot for financial literacy score by gender saved as 'financial_literacy_score_by_gender_boxplot.png'")
```

```python
numerical_cols = ['P4', 'TOTAL'] # P4 is age, TOTAL is the score

correlation_matrix = df_financial_literacy[numerical_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap of Financial Literacy Score and Age')
plt.savefig('financial_literacy_correlation_heatmap.png')
print("Heatmap for financial literacy correlation saved as 'financial_literacy_correlation_heatmap.png'")
# Financial Literacy Scores Analysis
# Score Distribution (Histogram)
# The histogram of financial literacy scores reveals the overall distribution of scores among the surveyed individuals. We can observe the range of scores and where most of the participants fall.

# Score by Gender (Box Plot)
# The box plot visualizes the financial literacy scores across different genders (Female, Male, Other). This allows us to compare the median scores, spread (interquartile range), and identify any outliers for each gender group.

# Feature Correlation (Heatmap)
# The heatmap displays the correlation between the numerical variables available in the dataset, specifically P4 (Age) and TOTAL (Financial Literacy Score). The correlation coefficient indicates the strength and direction of the linear relationship between these variables.

# Insights: Are teens financially literate? Who scores higher?
# Based on the generated plots:

# Overall Financial Literacy: To assess if teens are financially literate, we need to look at the Distribution of Financial Literacy Scores. If the scores are generally high (e.g., skewed towards the higher end of the possible score range), it would suggest a good level of financial literacy. Conversely, if scores are low or widely dispersed, it might indicate varying levels of understanding. Without a defined "pass mark" or a benchmark for what constitutes "financially literate," it's hard to give a definitive yes or no, but the distribution shows the general proficiency.

# Who Scores Higher?: By examining the Financial Literacy Score by Gender box plot:

# Compare the median lines within each box: The gender with a higher median line generally scores higher.

# Observe the spread of the boxes: A narrower box indicates more consistent scores within that gender group.

# Check for outliers: These are individual scores that fall significantly outside the main range for a particular gender.

# From the box plot, we can see that:

# Female individuals have a median score around 7.

# Male individuals have a median score around 7.

# Other individuals have a median score between 6 and 7, slightly lower than Female and Male.

# Overall, the median scores are very similar across Female and Male categories. The Other category appears to have a slightly lower median and a tighter distribution.

# Correlation with Age: The Correlation Heatmap shows the correlation coefficient between TOTAL (Financial Literacy Score) and P4 (Age). A positive correlation would suggest that as age increases, financial literacy scores tend to increase, and vice-versa for a negative correlation. A value close to 0 indicates little to no linear relationship. From the plot, the correlation between TOTAL and P4 is around 0.02, which is very close to zero. This suggests that in this dataset, there is very little to no linear correlation between age and financial literacy score. This implies that older teens are not necessarily more financially literate than younger teens based on this data.

# In summary: The financial literacy scores appear to be clustered around the middle range. There isn't a significant difference in median scores between males and females; both groups show similar central tendencies. The 'Other' gender category has a slightly lower median score. Additionally, age does not appear to be a strong predictor of financial literacy scores in this dataset.

```

```python
df_income_exp=pd.read_csv(r'Income_Expenditure_Data.csv')
# Create derived column: "Net Savings = Income  Expenses"
df_income_exp['Net_Savings'] = df_income_exp['Mthly_HH_Income'] - df_income_exp['Mthly_HH_Expense']
print("Derived column 'Net_Savings' created.")

# 1. Line plot: Income vs expenses by user (using DataFrame index as user/observation)
plt.figure(figsize=(12, 7))
plt.plot(df_income_exp.index, df_income_exp['Mthly_HH_Income'], label='Monthly Income', marker='o', linestyle='-')
plt.plot(df_income_exp.index, df_income_exp['Mthly_HH_Expense'], label='Monthly Expense', marker='x', linestyle='--')
plt.axhline(0, color='gray', linestyle='--', linewidth=0.8) # Add a horizontal line at 0 for reference
plt.title('Monthly Income vs. Monthly Expense per Household')
plt.xlabel('Household Index (User)')
plt.ylabel('Amount ($)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig('income_vs_expenses_line_plot.png')
print("Line plot for income vs. expenses saved as 'income_vs_expenses_line_plot.png'")

# Display a few rows with the new 'Net_Savings' column
print("\nFirst 5 rows with 'Net_Savings' column:")
print(df_income_exp.head().to_markdown(index=False, numalign="left", stralign="left"))
# Insights from Income & Savings Data
# Income vs. Expenses: By observing the Net_Savings column and the line plot, we can see that:

# Some households consistently have Mthly_HH_Income higher than Mthly_HH_Expense, indicating they are managing to save.

# Other households show Mthly_HH_Expense exceeding Mthly_HH_Income, resulting in negative savings (a deficit).

# The plot clearly illustrates the variability in financial stability across different households.

# Savings Presence: Direct 'savings' balance is not explicitly provided in this dataset. However, the Net_Savings column, which we derived, gives a clear indication of whether a household is accumulating savings or incurring debt on a monthly basis. This derived metric is crucial for understanding their financial health.

# Primary Sources of Income: The dataset includes Mthly_HH_Income and Annual_HH_Income, but does not provide a categorical breakdown of the sources of income (e.g., salary, business, investments, allowances). Therefore, a "Count plot: Source of income" as requested cannot be directly generated from the available columns. The No_of_Earning_Members column indicates how many individuals contribute to the household income, but not the nature of their income.

# Summary: The analysis of income and expenditure reveals a mixed financial landscape among the households. While some manage to save, others face deficits. The primary focus of the dataset is on the total income and expenditure rather than the specific breakdown of income sources.
```
